# 5-Number Summary and Box Plot

-> Minimum,Maximum,Median,Q1,Q2,IQR

```python
import numpy as np
```

```python
first_marks=[-14,1200,45,32,56,75,89,90,87,67,54,45,98,99,67,74,1000]
minimum,Q1,median,Q3,maximum=np.quantile(first_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
high_fence=Q3+1.5*(IQR)
print(lower_fence,high_fence)
```

```python
import seaborn as sns
```

```python
sns.boxplot(first_marks)
```
