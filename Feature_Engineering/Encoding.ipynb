# Data Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a simple dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

# One_Hot_Encoding

```python
#create an instance of OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

# Target Guided Ordinal Encoding

-> It is a technique used to encode categorical variable based on their relationship with the target variable. This encoding technique is useful when we have a categorical variable with a large number of unique categorical variable with a large number of unique categories and we want to use this variable as a feature in our machine learning model.

```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,200,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
df
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['price','city_encoded']]
```
